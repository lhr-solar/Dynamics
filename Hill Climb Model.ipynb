```python
import math
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import matplotlib.patches as mpatches
```

```python
# Equation 2.25 from The Winning Solar Car || Gives Force of gravitational pull up a hill
# Fg = W*sin(θ)
# W = 600 lbs according to Numerical Targets from Overall Folder in OneNote
angle = np.arange(0,math.pi/2,math.pi/240)
force = 600*np.sin(angle)
    
plt.plot(angle,force)
plt.title('Force vs Angle of Hill')
plt.xlabel('Angle (rads)')
plt.ylabel('Force(N)')
plt.grid(True,which='both')
plt.show()


# Equation 2.27 from The Winning Solar Car || Gives the Power use up a hill
# Pg = 1.99*W*V*
# W = 600 lbs
# V = 25 mph, 30 mph based on Numerical Targets from Overall Folder in OneNote
p25=(force*1.99*25)
plt.scatter(angle,p25)
l25 = mpatches.Patch(color = 'blue', label = '25 mph')
l30  = mpatches.Patch(color = 'orange', label = '30 mph')
first_legend = plt.legend(handles=[l25], loc = 'upper left')
ax = plt.gca().add_artist(first_legend)
second_legend = plt.legend(handles=[l30], loc = 'lower right')
p30=(force*1.99*30)
plt.scatter(angle,p30)
plt.title('Power vs Angle of Hill at Different Speeds')
plt.xlabel('Angle (rads)')
plt.ylabel('Power (W)')
plt.grid(True,which='both')
plt.show()
```
